I defined a function to help me read the data: read_train_test()

```python
folder_name = ['bending1', 'bending2', 'cycling', 'lying', 'sitting', 'standing', 'walking']
header_name = ['time', 'avg_rss12', 'var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23']
def read_train_test(folder_name_f, data_path):
    test_datas = pd.DataFrame([])
    train_datas = pd.DataFrame([])

    test_path = data_path + folder_name_f + '/' + 'test'
    file_list = os.listdir(test_path)
    for fl_te in file_list:
        read_path = test_path + '/' + fl_te
        temp_test_data = pd.read_csv(read_path, delimiter=',', skiprows=5, header=None)
        temp_test_data.columns = header_name
        test_datas = pd.concat([test_datas, temp_test_data])

    train_path = data_path + folder_name_f + '/' + 'train'
    file_list = os.listdir(train_path)
    # print(file_list)
    for fl_tr in file_list:
        # print(fl_tr)
        read_path = train_path + '/' + fl_tr
        # print(read_path)
        temp_train_data = pd.read_csv(read_path, delimiter=',', skiprows=5, header=None)
        # print(temp_train_data)
        # print(fl_tr)
        temp_train_data.columns = header_name
        train_datas = pd.concat([train_datas, temp_train_data])
    return test_datas, train_datas
```

### (c) Feature Extraction

#### i. Research

##### Time-domain features:
Mean: Average value \
Median: Middle value \
Mode: Most frequently occurring value \
Min, Max: min, max value \
Range: Max - min \
Variance: Quantifies the spread of the data \
Entropy: Measure the randomness/uncertainty in time series \
quartile: 1st, 2nd, 3rd quartile

#### ii. Extraction

```python
# We need first read all instances in

instances = []

# Next we need to read all data from each folders, here we need define several paths
paths = ['../AReM/bending1/', '../AReM/bending2/', '../AReM/cycling/', '../AReM/lying/', '../AReM/sitting/', '../AReM/standing/', '../AReM/walking/']

# Next we read each data from test/train.
for pt in paths:
    test_path = pt + 'test/'
    test_list = os.listdir(test_path)
    for tn in test_list:
        read_path = test_path + tn
        temp_test_data = pd.read_csv(read_path, delimiter=',', skiprows=5, header=None)
        instances.append(temp_test_data)
    # print(test_path)
    train_path = pt + 'train/'
    train_list = os.listdir(train_path)
    for tn in train_list:
        read_path = train_path + tn
        temp_train_data = pd.read_csv(read_path, delimiter=',', skiprows=5, header=None)
        instances.append(temp_train_data)

instances = np.array(instances)
print('Total read instances: ', instances.shape)
```

```python
# Next we extract the features

Features = []

for ins in range(0, instances.shape[0]):
    # Iterate through all instance
    # Get the instance
    ins_deal = instances[ins]
    # print(ins_deal.shape)

    # Split the data
    time_series_deal = np.hsplit(ins_deal, 7)
    # print(len(time_series_deal))
    for ts in range(1, 7):
        ts_deal = time_series_deal[ts]
        # print(ts_deal)
        ts_deal = np.array(ts_deal)
        Features.append([ts_deal.min(), ts_deal.max(), ts_deal.mean(), np.median(ts_deal), ts_deal.std(), np.percentile(ts_deal, 25), np.percentile(ts_deal, 75)])

Features = np.array(Features)
# print(Features.shape)
Features_re = Features.reshape(88, 6, 7)
# print(Features_re)
```

#### iii. Standard Deviation

```python
# Let we first get all standard deviation for each feature.
# Here we need to concat all datas for each feature.

instances_re = instances.reshape(42240, 7)
# print(instances_re.shape)
std_dev = []
for i in range(1, 7):
    std_dev.append(instances_re[:, i].std())

std_table = pd.DataFrame([std_dev[:]])
std_table.columns = ['avg_rss12', 'var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23']
std_table.index = ['STD_DEV']

std_table
```

```python
n_iterations = 1000
alpha = 0.1

std_dev_list = {hn: [] for hn in header_name[1:]}

instances_re_pd = pd.DataFrame(instances_re)
instances_re_pd.columns = header_name

# Iterate each time series
for hn in header_name[1:]:
    # Iterate for bootstrp
    for j in range(n_iterations):
        bootstrap = instances_re_pd[hn].sample(n = len(instances_re_pd), replace = True)
        value = bootstrap.std()
        # print(value)
        std_dev_list[hn].append(value)
        # print(bootstrap.std())
        # print(std_dev_list)
    # print(std_dev_list)

confidence_intervals = {}

for hn in header_name[1:]:
    lower = np.percentile(std_dev_list[hn], 100*alpha/2)
    upper = np.percentile(std_dev_list[hn], 100*(1-alpha/2))
    confidence_intervals[hn] = {'Lower': lower, 'Upper': upper}

print(pd.DataFrame(confidence_intervals).T)
```

### (a) Linear Train

For the training set, if we use a more flexible method, here is cubic regression, we will have a less RSS for training set. Because cubic regression will fit the training data more perfectly than linear.

### (b) Linear Test

However, cubic may perform very bad in testing set, so the linear regression will perform better.

### (c) Not Linear Train

For the training set, if we use a more flexible method, here is cubic regression, we will have a less RSS for training set. Because the true relationship is nonlienar, and cubic regression will fit the training data more perfectly than linear, also closer to real relationship.

### (d) Not Linear Testing

Cubic regression has more probability to perform better than linear, if it is not overfitting. However, if it is overfitting, it may perform worse. There is not enough information to tell it. Such as the true relationship is less than cubic, then face the same problem like (b).
